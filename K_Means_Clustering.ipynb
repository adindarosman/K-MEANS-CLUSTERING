```python
%matplotlib inline
from copy import deepcopy
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = (16,9)
plt.style.use('ggplot')
```

```python
#import dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
names = ['sepal-l','sepal-w','petal-l','petal-w','class']
data = pd.read_csv(url,names=names)
print (data.shape)
print (data.head(20))
```

```python
#plot dataset
f1 = data['sepal-l'].values
f2 = data['sepal-w'].values
f3 = data['petal-l'].values
f4 = data['petal-w'].values
X = np.array(list(zip(f1,f2,f3,f4)))
plt.scatter(f1, f2, c='black', s=7)
plt.scatter(f3, f4, c='black', s=7)
```

```python
from sklearn.cluster import KMeans
#menentukan jumlah cluster
kmeans = KMeans(n_clusters=4)
#fitting input data
kmeans = kmeans.fit(X)
#mendapatkan cluster labels
labels = kmeans.predict(X)
#mendapatkan nilai centroid
C = kmeans.cluster_centers_
#mencetak nilai centroid
print(C)
```

```python
plt.scatter(X[:, 0], X[:, 1], s=7, c=labels)
plt.scatter(C[:, 2], C[:, 3], marker='+', s=200, c='#050505')
```

```python
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import make_blobs
plt.rcParams['figure.figsize'] = (16, 9)
#men-generate dataset yang terkelompok dalam 4 cluster
X, y = make_blobs(n_samples=800, n_features=3, centers=4)
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:, 0], X[:, 1], X[:, 2])
```

```python
#initializing KMeans
kmeans = KMeans(n_clusters=4)
#fitting with inputs
kmeans = kmeans.fit(X)
#predicting the clusters
labels = kmeans.predict(X)
#getting the cluster centers
C = kmeans.cluster_centers_
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y)
ax.scatter(C[:, 0], C[:, 1], C[:, 2], marker='+', c='#050505', s=1000)
```

```python
for k in range (1, 10):
  #menentukan jumlah cluster
  kmeans = KMeans(n_clusters=k, random_state=1)
  #fitting input data
  kmeans = kmeans.fit(X)
  #mendapatkan cluster labels
  labels = kmeans.predict(X)
  #menghitung jumlahan jarak antara setiap sampel dengan cluster
  interia = kmeans.inertia_
  print ("k:",k, " cost:", interia)
print ("")
```

```python
from sklearn.metrics.cluster import silhouette_score
silhouette_score(X, labels)
```
